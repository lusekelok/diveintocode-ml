```python
def calculate_wheat_grain_number (n, m):
  board_list = [1]
  input_no = n * m          #to extract the index
  for index1 in range (input_no-1):   
    board_list.append(2*board_list[-1])     #creating a list of index
  board_ndarray = np.array(board_list)      #creating a numpy array from the list
  #print("The array is {}" .format(board_ndarray))
  board1_nda=board_ndarray.reshape(n,m)
  #print(board_ndarray.sum(axis=0))

  return board1_nda
calculate_wheat_grain_number (2, 2).sum(axis=0).sum(axis=0)
```

```python
#[Problem 5] How many times is the second half as long as the first half?

full_array=calculate_wheat_grain_number (8, 8)    #8by8 chessboard array
full_array
```

```python
#first half sum calculation

#print("The full array is{}" .format(full_array))
first_half=full_array[:4,:8]
print("The first half array is {}" .format(first_half))
firsthalf_sum =first_half.sum(axis=0).sum(axis=0)
print("The sum of first half is {}" .format(firsthalf_sum))
```

```python
#seond half sum calculation

second_half=full_array[4:8,:8]
print("The first half array is {}" .format(second_half))
secondhalf_sum=second_half.sum(axis=0).sum(axis=0)
print("The sum of second half is {}" .format(secondhalf_sum))
```

```python
# Difference between the first_half and second_half of 8by8 array

Diff_wheat_number = secondhalf_sum - firsthalf_sum
print("The difference between the wheat grain number between first half and second half of array is {}" .format(Diff_wheat_number))
```

```python
#[Problem 6] Expansion to n × m mass by another calculation method


def wheatsum_broadcast (n1, m1):    #using broadcast
  indices_of_squares1=0
  n_squares1 = 0
  n_squares1 = n1*m1
  indices_of_squares1 = np.arange(n_squares1).astype(np.uint64)
  board_ndarray1 = 2**indices_of_squares1
  board_ndarraysum1=board_ndarray1.sum(0).sum(0)
  return board_ndarraysum1
print(wheatsum_broadcast(8,8))
```

```python

def wheatsum_append(n2, m2):    #using append
  n_squares2 = n2*m2
  board_ndarray2 = np.array([1]).astype(np.uint64)
  for _ in range(n_squares2 - 1):
    board_ndarray2 = np.append(board_ndarray2, 2*board_ndarray2[-1])
    #board_res=board_ndarray2.reshape(n2,m2)
    board_ndarraysum2=board_ndarray2.sum(0)
  return board_ndarraysum2
print(wheatsum_append(8,8))
```
