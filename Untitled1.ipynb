```python
#Raising the power code
def raise_to_the_power(base, power):
  pow=1
  for i in range(1, power + 1):
    pow =pow * base   
  return pow
```

```python
##[Problem 1] Number of rice grains on the 100th day

GRAINS = 123
test_list = []
test_list1 = []
test_list2 = []
i = 0
for i in range(GRAINS):
      test_list.append(i)
print (test_list)   #just to see what is been stored
x = 0
index = 0
for index in test_list:
  x = raise_to_the_power(2,index)
  test_list1.append(x) 
print (x)          #number of rice grains on 100th day
#test_list1.append(x)
print (test_list1)    #just to see what is been stored
print (sum(test_list1))   #number of rice grains from 1st day upto by 100th day
y = 0
for index1 in test_list1:
  y = y + index1
  test_list2.append(y)
print(test_list2)     #cummulative number of rice grains upto each day
```

```python
#Graphing the lines
import matplotlib.pyplot as plt
%matplotlib inline

# line 1 points
x1 = test_list
y1 = test_list1
# plotting the line 1 points 
plt.plot(x1, y1, label = "Rice Grain on each day")
  
# line 2 points
x2 = test_list
y2 = test_list2
# plotting the line 2 points 
plt.plot(x2, y2, label = "Rice Grain Cummulative")
  
# naming the x axis
plt.xlabel('number of days')
# naming the y axis
plt.ylabel('number of rice')
# giving a title to my graph
plt.title('Two lines on same graph!')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```
