```python
#Creating dummy data

import numpy as np
np.random.seed(0)
mean = np.array([-3, 0])
print(mean)
matrix = np.array([1.0, 0.8, 0.8, 1.0])
#print(matrix)
cov_matrix = matrix.reshape(2,2)
print(cov_matrix)
multi_var = np.random.multivariate_normal(mean, cov_matrix, size=(500,2))
print(multi_var)
x1 = multi_var[:500, :1]
print(x1.shape)
print(x1.ndim)
```

```python
#[Problem 2] Visualization by scatter plot

import matplotlib.pyplot as plt
%matplotlib inline
x3 = multi_var[:500, :1]
#print(x3)
x4=x3.reshape(500,2)
#print(x4)
x = x4[:500,:1]    #extracting the X-axis data
#print(x5)
y = x4[:500,1:]    #extracting the Y-axix data
#print(x6)
plt.scatter(x,y,s=100,c="blue", edgecolors="black",linewidths=1,alpha=0.75)    #plotting the X and Y on scatterplot
plt.xlabel("X1")
plt.ylabel("x2")
plt.title("scatter")
```
