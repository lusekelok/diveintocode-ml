```python
import numpy as np
import pandas as pd
```

```python
#[Problem 1] Data acquisition

from sklearn.datasets import load_iris    #load iris file
dir(load_iris())    #to view list of attributes
```

```python
iris = load_iris()    #storing loaded iris data in an object called iris
print (iris.feature_names)
```

```python
X = iris.data       #assigning the iris data to X
print(X)
y = iris.target     #assigning the iris target to y
print(y)
```

```python
#[Problem 2] Combining data
y1 = y.reshape(150,1)
df = np.hstack((X, y1))
print(df)
```

```python
#print(df.shape)
iris.feature_names.append("Species")    #appending a Species column to match the number of columns of feature_names with the created df
iris.feature_names
```

```python
#data = load_iris()
df1 = pd.DataFrame(df, columns=iris.feature_names)
df1.head()
```

```python
#[Problem 3] Checking the data

#Display the 4th sample from the beginning and see what each feature is like. (Numerical data or character data, etc.)
df1.iloc[[4]]
```

```python
#[Problem 5] Extracting the required data

#Extractsepal_width, columns in three different ways
SW = df1.iloc[:,1:2]  #extracting sepal width using the iloc[] function by using indices
print(SW)

SW1 = df1[df1.columns[1]]  #extracting sepal width by specifying column number
print(SW1)

SW2 = df1[["sepal width (cm)"]]   #extraction by column name wise
print(SW2)
```

```python
#[Problem 6] Creating a diagram

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Setosa', 'virgiclor', 'virginica'
sizes = [33.3, 33.3, 33.3]
explode = (0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
